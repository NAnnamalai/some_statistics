```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(color_codes=True)
```

```python
# Dataset: http://guidetodatamining.com/chapter2/
df = pd.read_csv("Movie_Ratings.csv", encoding="cp1252")
df.head()
```

```python
df = df.rename(columns={'Unnamed: 0': 'Movies'})
df.head()
```

```python
df['Movies'].nunique(), len(df.columns)
```

```python
data = df.copy()
data = data.set_index('Movies')
statistics_df = pd.DataFrame()
statistics_df['sum'] = data.sum(axis=1)
statistics_df['mean'] = data.mean(axis=1)
```

```python
statistics_df['median'] = data.median(axis=1)
```

```python
statistics_df['mode_first_term'] = data.mode(axis=1)[0]
statistics_df['mode_second_term'] = data.mode(axis=1)[1]
```

```python
statistics_df['ratings_range'] = data.max(axis=1) - data.min(axis=1)
```

```python
statistics_df['var'] = data.var(axis=1)
```

```python
statistics_df['std'] = data.std(axis=1)
```

```python
statistics_df['75th_quantile'] = data.T.quantile(0.75)
```

```python
statistics_df['25th_quantile'] = data.T.quantile(0.25)
```

```python
statistics_df['skewness'] = data.skew(axis=1)
```

```python
sns.set(font_scale=1.4)
sns.set(rc={'figure.figsize':(14, 11)})
plt.title("Visualizing Total No. Of Rating For Each Movie", y=1.02);
tot_df = statistics_df.sort_values('sum', ascending=False)
ax = sns.barplot(y=tot_df.index, x='sum', data=tot_df, color="salmon", saturation=.9, ci="sd")
```

```python
sns.set(font_scale=1.4)
sns.set(rc={'figure.figsize':(14, 11)})
plt.title("Visualizing Skewness", y=1.02);
skew_df = statistics_df.sort_values('skewness', ascending=False)
ax = sns.barplot(y=skew_df.index, x='skewness', data=skew_df, color="salmon", saturation=.9)
```

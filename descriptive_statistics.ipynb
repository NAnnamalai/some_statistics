```python
import pandas as pd
```

```python
# Dataset: http://guidetodatamining.com/chapter2/
df = pd.read_csv("Movie_Ratings.csv", encoding="cp1252")
df.head()
```

```python
df = df.rename(columns={'Unnamed: 0': 'Movies'})
df.head()
```

```python
df['Movies'].nunique(), len(df.columns)
```

```python
data = df.copy()
data = data.set_index('Movies')
statistics_df = pd.DataFrame()
statistics_df['mean'] = data.mean(axis=1)
```

```python
statistics_df['median'] = data.median(axis=1)
```

```python
statistics_df['mode_first_term'] = data.mode(axis=1)[0]
statistics_df['mode_second_term'] = data.mode(axis=1)[1]
```

```python
statistics_df['ratings_range'] = data.max(axis=1) - data.min(axis=1)
```

```python
statistics_df['var'] = data.var(axis=1)
```

```python
statistics_df['std'] = data.std(axis=1)
```

```python
statistics_df['75th_quantile'] = data.T.quantile(0.75)
```

```python
statistics_df['25th_quantile'] = data.T.quantile(0.25)
```

```python
statistics_df['skewness'] = data.skew(axis=1)
```
